# Transforming Poisson to Normal Distribution

```python
sample_for_poisson = sts.poisson.rvs(mu = 2.5, size = 500)
poisson_transformed_to_normal,lmbda = sts.boxcox(sample_for_poisson+1)
print("Lambda - ",lmbda)
```

```python
plt.hist(poisson_transformed_to_normal)
```

```python
sns.kdeplot(poisson_transformed_to_normal)
```

```python
# Let's do Anderson Darling Test to check if it's normal or not
```

```python
sts.anderson(poisson_transformed_to_normal,dist='norm')
```

```python
# Conclusion - Even at 1% significance level i can reject the null hypothesis
# Reason why we are getting this multiple hills is after transforming the data is still dicrete
```

```python
# Conclusion - Poisson distribution can not be converted into Normal distribution by box-cox transformation, Practically.
# because discrete rv to continuous rv transformation is possible with box-cox transformation but it does not
# make any sense
# If we want to convert poisson distributioin to normal then we have to use something called Ancombe's Transform.
```

# Why mode of exponential dist at 0?

```python
#An exponential distribution is that of a continuous random variable.
# All particular values it can take have probability mass of zero.
#The mode of a continuous random variable is not the point where its probability is most massive.
# It is where it is most dense.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller_transform(u1, u2):
    # Box-Muller transform to generate two independent normal random variables
    z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    return z1, z2

def generate_normal_samples(mean, std_dev, sample_size):
    # Generate uniform random numbers between 0 and 1
    u1 = np.random.uniform(0, 1, sample_size // 2)
    u2 = np.random.uniform(0, 1, sample_size // 2)

    # Apply Box-Muller transform to generate standard normal variables
    z1, z2 = box_muller_transform(u1, u2)

    # Scale and shift to obtain samples from desired normal distribution
    samples = np.concatenate((z1, z2)) * std_dev + mean
    return samples

# Parameters for the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
sample_size = 1000  # Number of samples to generate

# Generate random samples from a normal distribution
samples = generate_normal_samples(mean, std_dev, sample_size)

# Plotting the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Adding a normal distribution curve for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi * std_dev**2))) * np.exp(- (x - mean)**2 / (2 * std_dev**2))
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram of Random Samples from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```
